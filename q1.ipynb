### **Item 1.a**
Mostre que, para $Q_vx = \|x\|e_1$, temos $v = x - \|x\|e_1$ e $\beta = \frac{1}{\|x\|(\|x\|- x_1)} = \frac{\|x\|+x_1}{\|x\|(\|x\|^2-x^2_1)} = \frac{\|x\|+x_1}{\|x\|\cdot\|y\|^2}$, onde $y = (x_2,...,x_n)$.

Seja $x \in \mathbb{R}^n$, $v \in \mathbb{R}^n$  e $\beta \in \mathbb{R}$, queremos determinar $v$ e $\beta$ tais que $Q_vx = (I - \beta vv^*)x = \|x\|e_1$. Logo, considerando $x = (x_1, x_2, ..., x_n)$ temos que:
$$
\begin{aligned}
Q_vx = (I - \beta vv^*)x &= \|x\|e_1
\\
x - \beta (v^*x) v &= \|x\|e_1
\\
x_1 - \beta (v^*x) v_1 &= \|x\| 
\\
\text{e } x_i - \beta (v^*x) v_i &= 0, \text{ para } i=2,..., n 
\end{aligned}
$$

Com isso temos que 
$$
\begin{aligned}
v_1 &= \frac{x_1 - \|x\|}{\beta (v^*x)}
\\
v_i &= \frac{x_i}{\beta (v^*x)},  \text{ para } i=2,...,n
\end{aligned}
$$

Dado que $\beta (v^*x)$ é uma constante, temos que as entradas de $v$ são proporcionais à $x - \|x\|e_1$. Agora, note que multiplicar $v$ por um escalar continua definindo o mesmo refletor se esse escalar for absorvido em $\beta$. Logo, podemos tomar $\beta (v^*x) = 1$, chegando em $v = x - \|x\|e_1$.

Agora, temos que 
$$
\beta (v^*x) = 1 \Rightarrow \beta = \frac{1}{v^*x} = \frac{1}{(x - \|x\|e_1)^*x} = \frac{1}{\|x\|^2 - \|x\|x_1} = \frac{1}{\|x\|(\|x\| - x_1)}
$$

E com algumas manipulações alcançamos as outras fórmulas para $\beta$
$$
\begin{aligned}
\beta_1 &= \frac{1}{\|x\|(\|x\| - x_1)}
\\
\beta_2 &= \beta_1 \cdot \frac{\|x\| + x_1}{\|x\| + x_1} = \frac{\|x\| + x_1}{\|x\|(\|x\|^2 - x_1^2)}
\\
\beta_3 &= \beta_2 = \frac{\|x\|+x_1}{\|x\|\sum_{i=2}^n{x_i^2}} = \frac{\|x\|+x_1}{\|x\|\|y\|^2}, \text{onde } y=(x_2, x_3,..., x_n) \in \mathbb{R}^{n-1}
\end{aligned}
$$

### **Item 1.b**
Verifique que, dependendo do sinal de $x_1$, uma das fórmulas acima para $\beta$ não está sujeita a perda de precisão devido a cancelamento numérico.

Sejam $\beta_1, \beta_2, \beta_3$ as três fórmulas para $\beta$. Note que, tanto $\beta_1$ quanto $\beta_2$ possuem um termo $\|x\| - x_1$ no denominador. Esse termo é problemático no quesito de cancelamento numérico pois, quando $x$ é, aprximadamente, $\alpha e_1$, ou seja, quase colinear com $e_1$, temos que $|x_1| \approx \|x\|$, e, se $x_1 \ge 0$, o termo $\|x\| - x_1\approx 0$. 

Ou seja, quando $x_1 \approx \|x\| \ge 0$, as fóromulas $\beta_1$ e $\beta_2$ são altamente instável devido a esse termo muito pequeno no denominador. Por outro lado $\beta_3$ não realiza o cálculo de $\|x\| - x_1$, de forma que esse termo extremamente pequeno no denominador não apareça, evitando o cancelamento e, portanto, sendo mais estável.

Agora, se $x_1 \approx -\|x\|$, o termo $\|x\| + x_1$ passa a se tornar númericamente instável, pois $\|x\| + x_1 \approx \|x\| - \|x\| = 0$. Ou seja, como as fórmulas $\beta_2$ e $\beta_3$, que possuem esse termo, passam a perder precisão devido a esse cancelamento. Por outro lado, a fórmula $\beta_1$ não realiza esse cálculo, evitando perda de precisão, e o termo $\|x\| - x_1 \approx \|x\| + \|x\|$ se torna uma soma de valores positivos, de forma que não cancele.

Ou seja, se $x_1 \ge 0$, as fórmulas $\beta_1$ e $\beta_2$ são instáveis devido à possibilidade de cancelamento no termo $\|x\| - x_1 \approx 0$, enquanto a fórmula $\beta_3$ não realiza esse cálculo, mantendo sua precisão.
Por outro lado, se $x_1 < 0$, as fórmulas $\beta_2$ e $\beta_3$ são instáveis devido ao termo $\|x\| + x_1 \approx 0$, enquanto a fórmula $\beta_1$ não realiza esse cálculo, mantendo sua precisão.

Também vale notar que, quando $x_1 \approx \|x\|$, o cálculo de $v$ também corre risco de perder precisão devido à cancelamento numérico. Logo, quando $x_1 \ge 0$, podemos nos inspirar na fórmula $\beta_3$ e calcular $v_1$ da seguinte forma:

$$
v_1 = x_1 - \|x\| = (x_1 - \|x\|) \cdot\frac{x_1 + \|x\|}{x_1 + \|x\|} = \frac{x_1^2 - \|x\|^2}{x_1 + \|x\|} = \frac{-\|y\|^2}{x_1 + \|x\|}
$$

O impacto dessa alteração será dicutido no [item 1.g](#item-1g).

### **Item 1.c**
Escreva uma função ```reflector(x)``` que, dado um vetor $x$, calcula o vetor $v$ e o escalar $\beta$ do refletor de Householder $Q$ tal que $Qx = \|x\|e_1$.

````python
import numpy as np

def reflector(x):
	x = np.asarray(x)

	e1 = np.zeros(x.shape[0], dtype=x.dtype)
	e1[0] = 1

	x_norm = np.linalg.norm(x)
	y_norm2 = np.sum(x[1:] ** 2)

	v = x.copy()

	if y_norm2 <= 1e-16:
		return v, 0.0

	if x[0] >= 0:
		v[0] = - y_norm2 / (x_norm + x[0]) 
		beta = (x_norm + x[0]) / (x_norm * y_norm2)

	else:
		v[0] -= x_norm
		beta = 1 / (x_norm * (x_norm - x[0]))

	return v, beta
````

### **Item 1.d**
**Condicionamento:** Mostre que a derivada de $v$ em relação a $x$ é $J = I -\frac{e_1x^∗}{\|x\|}$. Deduza que o número de condicionamento absoluto de $v$ em relação a $x$ é menor ou igual a $2$.

Como vimos no [item 1.a](#item-1a), $v = x - \|x\|e_1$, ou seja:
$$
v = x - \|x\|e_1 \iff
\begin{cases}
v_1 = x_1 - \|x\|
\\
v_2 = x_2
\\
v_3 = x_3
\\
\vdots
\\
v_n = x_n
\end{cases}
$$

Logo, cqlculando a matriz $J(x)$, temos que a partir da sua segunda linha, a diagonal principal terá $1$, enquanto o resto será $0$. A primeira linha, por outro lado, para a primeira linha, é necessário avaliar a norma $\|x\| = \sqrt{x_1^2 + x_2^2 + ... + x_n^2}$. Logo, teremos que:

$$
\begin{aligned}
\frac{\partial v_1}{\partial x_1} &= 1 - \frac{x_1}{\sqrt{x_1^2 + x_2^2 + ... + x_n^2}} = 1 - \frac{x_1}{\|x\|}
\\
\frac{\partial v_1}{\partial x_i} &= \frac{-x_i}{\sqrt{x_1^2 + x_2^2 + ... + x_n^2}} = \frac{-x_i}{\|x\|}, i = 2,...,n
\end{aligned}
$$

Agora, se olharmos para $\frac{-e_1x^*}{\|x\|}$, veremos que esse termo gera uma matriz $n\times n$, tal que os termos $\frac{-x_i}{\|x\|}$ ocupem a primeira linha. Subitraindo da identidade, obtemos exatamente as derivadas parciais nas posições corretas.

$$
J(x) = 
\begin{bmatrix}
1 - \frac{x_1}{\|x\|} & \frac{-x_2}{\|x\|} & \frac{-x_3}{\|x\|} & ... & \frac{-x_n}{\|x\|}
\\
0 & 1 & 0 & ... & 0
\\
0 & 0 & 1 & ... & 0
\\
\vdots & \vdots & \vdots & \ddots & \vdots
\\
0 & 0 & 0 & ... & 1
\end{bmatrix}
$$

Quando ao condicionamento, queremos mostrar que o condicionamento $k$ da matriz $J(x)$ é menor que $2$ ou seja, que $max_{x\neq y \in \mathbb{R}^n} \frac{\|J(x) - J(y)\|}{\|x-y\|}\le 2$. Entretanto, conforme vimos em aula, o condicionamente absoluto da transformação $x\to v$ é igual a norma da matriz $J(x)$, logo, podemos demostrar isso apresentando cotas superiores para essa norma:

$$
\|J(x)\| = \| I - \frac{e_1x^*}{\|x\|}\| \le \|I\| + \|\frac{e_1x^*}{\|x\|}\|, \text{ desigualdade triangular}
$$

Sabe-se que a matriz $I$ possui norma $1$, enquando $e_1x^*$ tem norma igual $\|e_1\|\cdot\|x\| = \|x\|$, em vista de ser uma matriz de posto $1$ definida pelo produto vetorial desses dois vetores. Logo, temos que:

$$
\|J(x)\| \le \|I\| + \|\frac{e_1x^*}{\|x\|}\| = 1 + \frac{\|e_1\|\cdot\|x\|}{\|x\|} = 1 + 1 = 2
$$

Para estudar o efeito do cancelamento númerico observado no [item 1.b](#item-1b), construímos uma família de vetores $x(s)$ que varia continuamente entre um vetor genérico (para $s=1$) e o vetor $\|x\|e_1$ (para $s \to 0$). A construção é feita fixando a norma $\|x(s)\| = \|x\|$ constante e escalando as componentes "laterais" por s, de modo que:
$$
y(s) = s \cdot y_0, \quad x(s) = \left(\sqrt{\|x\|^2 - \|y(s)\|^2},\ y(s)\right)
$$
À medida que $s \to 0$, $x(s) \to \|x\|e_1$​, e é exatamente nesse limite que o cancelamento numérico entre $x_1​$ e $\|x\|$ se torna crítico. Amostramos os valores de $s$ em escala logarítmica de $1$ até $10^{-14}$, e tomamos os cálculos em `Float64` como referência de "verdade" para o cálculo.

````python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

def calc_trajectoria(x0, amostras, limite=14):
    x0 = np.asarray(x0, dtype=np.float64)
    r = np.linalg.norm(x0)
    y0 = x0[1:]

    s_vals = np.logspace(0, -limite, amostras)  # de 1 até 10^-limite

    xs = []
    for s in s_vals:
        y = s * y0
        x1 = np.sqrt(max(r**2 - np.dot(y, y), 0.0))
        x = np.concatenate(([x1], y))
        xs.append(x)

    return xs, s_vals

def calc_vbeta(xs, dtype_a, dtype_b):
    results = []
    for x in xs:
        xA = x.astype(dtype_a)
        xB = x.astype(dtype_b)

        vA, betaA = reflector(xA)
        vB, betaB = reflector(xB)

        results.append({'vA':vA.astype(np.float64), 'betaA':float(betaA),
                        'vB':vB.astype(np.float64), 'betaB':float(betaB),
                        'x':x.astype(np.float64)})
    return results

````

````python
x0 = np.random.randint(1000.0, size=10)
xs, ss = calc_trajectoria(x0, 100, 14)
results = calc_vbeta(xs, np.float32, np.float64)
````

````python
r = pd.DataFrame(results)

b_32 = r['betaA'].values
b_64 = r['betaB'].values
diff_b = np.maximum(np.abs(b_32 - b_64) / np.abs(b_64)  , 1e-16)

fig, axs = plt.subplots(1, 2, figsize=(12, 6))

sns.lineplot(x=ss, y=diff_b, label=r"$\frac{|\beta_{32}-\beta_{64}|}{|\beta_{64}|}$", ax=axs[0])
axs[0].invert_xaxis()
axs[0].set_xscale('log')
axs[0].set_yscale('log')
axs[0].set_xlabel(r"$s$ - proximidade $\|x\|e_1$ conforme $s\to 0 $")
axs[0].set_ylabel(r"erro relativo de $\beta$")
axs[0].set_title(r"Estabilidade de $\beta$: `Float32` vs `Float64`")

diff_v = np.array([np.maximum(np.linalg.norm(r['vA'] - r['vB']) / np.linalg.norm(r['vB'])  , 1e-16) for r in results])

sns.lineplot(x=ss, y=diff_v, label=r"$\frac{\|v_{32}-v_{64}\|}{\|v_{64}\|}$", ax=axs[1])
axs[1].invert_xaxis()
axs[1].set_xscale('log')
axs[1].set_yscale('log')
axs[1].set_xlabel(r"$s$ - proximidade $\|x\|e_1$ conforme $s\to 0 $")
axs[1].set_ylabel(r"erro relativo de $v$")
axs[1].set_title(r"Estabilidade de $v$: `Float32` vs `Float64`")

plt.tight_layout()
plt.show()
````

Conforme discutido no [item 1.b](#item-1b), a escolha da formula correta para o cálculo de $\beta$ e de $v$ permite a minimização da perda de precisão por cancelamento númerico. Os gráficos demostram exatamente isso, apesar de $x(s)$ se aproximar cada vez mais de $\|x\|e_1$ conforme $s\to 0$, o erro relativo continua na mesma ordem da precisão da máquina com `Float32` ($~10^{-8}$), de modo que a implementação seja numericamente estável.

No [item 1.g](#item-1g) veremos o impacto de não usar as fórmulas apresentadas no [item 1.b](#item-1b).

````python
def calc_residuo(r, dtype):
    x = r['x'].astype(dtype)
    v = (r['vA'] if dtype == np.float32 else r['vB']).astype(dtype)
    beta = dtype(r['betaA'] if dtype == np.float32 else r['betaB'])

    norma = np.linalg.norm(x)
    e1 = np.zeros(x.shape[0], dtype=dtype)
    e1[0] = 1

    Qvx = x - beta * (v @ x) * v
    return float(np.linalg.norm(Qvx - norma * e1))

res64 = np.maximum([calc_residuo(r, np.float64) for r in results], 1e-16)
res32 = np.maximum([calc_residuo(r, np.float32) for r in results], 1e-16)

sns.lineplot(x=ss, y=res64, label=r"`Float64`")
sns.lineplot(x=ss, y=res32, label=r"`Float32`")
plt.gca().invert_xaxis()
plt.xscale('log')
plt.yscale('log')
plt.xlabel(r"$s$ - proximidade $\|x\|e_1$ conforme $s\to 0 $")
plt.ylabel(r"$\|Q_v x - \|x\|e_1\|$")
plt.title(r"Resíduo da reflexão")
plt.tight_layout()
plt.show()
````

O gráfico apresenta um comportamento diferente do que se imaginaria com base nas análises anteriores. O comportamento do `Float32` é particularmente interessante, para $s \approx 1$ (vetor genérico), o resíduo da reflexão em `Float32` é grande (na ordem de $~10^{-3}$), o que indica que a operação $Q_v x \approx \|x\|e_1$​ tem erro absoluto considerável nessa precisão. Isso não é efeito de cancelamento, como vimos anteriormente, mas apresenta uma limitação do próprio `Float32` ao operar com vetores de componentes de magnitudes diversas. Conforme $s$ diminui e $x$ se alinha cada vez mais com $e_1$​, a reflexão se torna "mais fácil" numericamente, pis o vetor já está quase no lugar certo, e o resíduo cai. Por outro lado, `Float64` mantém resíduos próximos de sua precisão de máquina ($~10^{-15}$) em quase todo o intervalo.

````python
dims   = [2, 3, 5, 8, 12, 20, 35, 50, 75, 100]   
n_s    = 30                 
s_vals = np.logspace(0, -14, n_s) 

np.random.seed(42)

grid_beta = np.full((len(dims), n_s), np.nan)
grid_v    = np.full((len(dims), n_s), np.nan)

for i, d in enumerate(dims):
    # vetor base fixo para esta dimensão
    x0    = np.random.randn(d)
    x0   /= np.linalg.norm(x0)
    r     = 1.0
    y0    = x0[1:]

    for j, s in enumerate(s_vals):
        y   = s * y0
        x1  = np.sqrt(max(r**2 - np.dot(y, y), 0.0))
        x   = np.concatenate(([x1], y))

        v32, b32 = reflector(x.astype(np.float32))
        v64, b64 = reflector(x.astype(np.float64))

        b32f, b64f = float(b32), float(b64)
        err_beta = abs(b32f - b64f) / abs(b64f) if b64f != 0 else np.nan
        err_v    = (np.linalg.norm(v32.astype(np.float64) - v64)
                    / np.linalg.norm(v64)) if np.linalg.norm(v64) > 0 else np.nan

        grid_beta[i, j] = np.log10(max(err_beta, 1e-16))
        grid_v[i, j]    = np.log10(max(err_v,    1e-16))


col_labels = [f"1e{int(np.log10(s)):+d}" if s < 1 else "1e+0"
              for s in s_vals]
row_labels = [str(d) for d in dims]

df_beta = pd.DataFrame(grid_beta[::-1],
                        index=row_labels[::-1],
                        columns=col_labels)

df_v    = pd.DataFrame(grid_v[::-1],
                        index=row_labels[::-1],
                        columns=col_labels)

fig, axes = plt.subplots(2, 1, figsize=(14, 9))

common = dict(
    cmap="RdYlGn_r",   # verde = erro baixo, vermelho = erro alto
    vmin=-16, vmax=-4,
    linewidths=0,
    xticklabels=4,     # mostra 1 a cada 4 labels no eixo x
    cbar_kws={"label": "log₁₀(erro relativo)", "shrink": 0.8},
)

for ax, df, title in zip(axes,
                          [df_beta, df_v],
                          [r"Erro relativo de $\beta$  (log₁₀)",
                           r"Erro relativo de $v$  (log₁₀)"]):
    sns.heatmap(df, ax=ax, **common)
    ax.set_title(title, fontsize=13, pad=8)
    ax.set_xlabel(r"$s$  (de $1$ → $10^{-14}$  da esquerda para a direita)", fontsize=11)
    ax.set_ylabel("dimensão de $x$", fontsize=11)
    ax.tick_params(axis='x', rotation=45, labelsize=8)
    ax.tick_params(axis='y', rotation=0,  labelsize=9)

    ax.axvline(x=n_s * 0.5, color='white', lw=0.8, linestyle='--', alpha=0.4)

plt.suptitle("Grid Search — Estabilidade do refletor: Float32 vs Float64\n"
             "Cor: log₁₀(erro relativo) — Verde ≈ precisão de máquina, Vermelho = instabilidade",
             fontsize=12, y=1.01)
plt.tight_layout()
plt.show()
````

O gráfico acima mostra que esse comportamento se reproduz independente do $x$ específico. Confirmando o que discutimos antes.

### Item 1.f
Escreva uma função ```calc_beta(v)``` que calcula $\beta$ unicamente a partir de $v$. Repita os testes ```Float64``` e ```Float32``` usando esta função, e compare os resultados com a função anterior. Observe que não há muita diferença ao calcular $\beta$ a partir de $x$ ou apenas de $v$, mesmo com $x$ quase alinhado com $e_1$.

Vimos em aula que um refletor de Householder $Q_v$ é dado da forma $Q_v = I - 2\frac{vv^*}{v^*v}$, enquanto a forma $Q_v = I - \beta vv^*$ é para refletores mais genéricos. Com base nisso (e inspirações de conversas com colegas), contruiremos a demostração de que $\beta = \frac{1}{v^*x} = \frac{2}{v^*v} \Rightarrow Q_v x = \|x\|e_1$.

$$
\begin{aligned}
\beta &=\frac{1}{v^*x} = \frac{2}{v^*v} 
\\
&\iff v^*v = 2 v^*x
\\
&\iff v^*(v - 2x) = 0 
\end{aligned}
$$

Sabemos pelo [item 1.a](#item-1a) que, para um refletor $Q_v$, $v = (x_1 - \|x\|, x_2, x_3, ..., x_n)$, agora, basta verificar que $v^*(v - 2x) = 0$ para o vetor $v$ desse refletor:

$$
v - 2x = (-x_1 -\|x\|, -x_2, -x_3, ..., -x_n)
\\
\Rightarrow v^*(v-2x) = -(x_1 - \|x\|)(x_1 + \|x\|) - \sum_{i=2}^n{x_i^2} = \|x\|^2 - x_1^2 - \sum_{i=2}^n{x_i^2} = \|x\|^2 - \|x\|^2 = 0
$$

Logo, temos uma fórmula dependente apenas de $v$ para $\beta$.

````python
def calc_beta(v):
    return 2 / (v @ v)

# ctrl c do teste de tipos anterior
x_f32 = np.array([1e10, 0, 0, 0, 1], dtype=np.float32)
x_f64 = np.array([1e10, 0, 0, 0, 1], dtype=np.float64)

e1 = np.zeros_like(x_f32)
e1[0] = 1

b_f32 = calc_beta(x_f32 - np.linalg.norm(x_f32) * e1)
b_f64 = calc_beta(x_f64 - np.linalg.norm(x_f64) * e1)

print("Tipos Resultantes:")
print(f"x: {x_f32.dtype} - beta: {type(b_f32).__name__}")
print(f"x: {x_f64.dtype} - beta: {type(b_f64).__name__}")
````

````python
def calc_2_betas(x):
	if not isinstance(x, np.ndarray):
		raise ValueError('x must be an np.array')

	e1 = np.zeros(x.shape[0], dtype=x.dtype)
	e1[0] = 1

	x_norm = np.linalg.norm(x)
	y_norm2 = np.sum(x[1:] ** 2)

	v = x.copy()

	if x[0] >= 0:
		v[0] = - y_norm2 / (x_norm + x[0]) 
		betaA = (x_norm + x[0]) / (x_norm * y_norm2)

	else:
		v[0] -= x_norm
		betaA = 1 / (x_norm * (x_norm - x[0]))

	betaB = 2 / (v @ v)

	return betaA, betaB


def compara_betas(xs, dtype_a, dtype_b):
    results = []
    for x in xs:
        xA = x.astype(dtype_a)
        xB = x.astype(dtype_b)

        beta_xA, beta_vA = calc_2_betas(xA)
        beta_xB, beta_vB = calc_2_betas(xB)

        results.append({'beta_xA':float(beta_xA), 'beta_vA':float(beta_vA),
                        'beta_xB':float(beta_xB), 'beta_vB':float(beta_vB),
                        })
    return results
````

### Item 1.g

Mostre/Verifique que $v_1$ é (relativamente!) mal-condicionado quando $x$ está quase alinhado com $e_1$. Inspirado na fórmula para $\beta$, proponha outra fórmula para calcular $v_1$ (ainda refletindo $x$ em $\|x\|e_1$) que não seja tão sensível a $x_1$. 

Obs: O objetivo de refletir um vetor com $x_1 > 0$ em um vetor com a primeira coordenada positiva é garantir que a matriz $R$ terá elementos positivos na diagonal.

Vimos no [item 1.d](#item-1d) que o condicionamento absoluto de $v$ em relação a $x$ é, no máximo, $2$, ou seja, $v$ como um todo é bem-condicionado. Entretanto, isso não significa que cada componente de $v$ individualmente seja bem-condicionado relativamente. Em particular, analisamos $v_1 = x_1 - \|x\|$.

Quando $x$ está quase alinhado com $e_1$ (ou seja, $x_1 \approx \|x\|$, com $x_1 \ge 0$), temos $v_1 \approx 0$. Isso é exatamente o cancelamento discutido no [item 1.b](#item-1b): estamos subtraindo dois números muito próximos, $x_1$ e $\|x\|$, e o resultado $v_1$ é muito pequeno em valor absoluto. O problema é que, embora o erro absoluto em $v_1$ permaneça da ordem do erro de arredondamento de $x_1$ e $\|x\|$ (tipicamente $\sim\epsilon_{\text{máquina}}\|x\|$), o erro relativo de $v_1$ explode, pois
$$
\text{erro relativo de } v_1 \sim \frac{\epsilon_{\text{máquina}}\|x\|}{|v_1|} \to \infty \quad \text{quando } v_1 \to 0.
$$
Ou seja, $v_1$ é relativamente mal-condicionado nesse regime: poucos algarismos significativos corretos sobram no resultado, mesmo que o erro absoluto seja pequeno.

Esse é precisamente o motivo pelo qual, no [item 1.b](#item-1b), propusemos a fórmula alternativa
$$
v_1 = \frac{-\|y\|^2}{x_1 + \|x\|}, \qquad y = (x_2, ..., x_n),
$$
obtida multiplicando e dividindo $x_1 - \|x\|$ por $x_1 + \|x\|$. Essa expressão não envolve a subtração de dois números quase iguais: o numerador $-\|y\|^2$ é calculado diretamente a partir das componentes "pequenas" de $x$ (que carregam toda a informação relevante quando $x$ está quase alinhado com $e_1$), e o denominador $x_1 + \|x\|$ é uma soma de dois números positivos (quando $x_1 \ge 0$), portanto bem-condicionada. Com isso, evitamos o cancelamento e $v_1$ passa a ser calculado com erro relativo da ordem da precisão da máquina, independentemente de quão próximo $x$ esteja de $e_1$.

````python
def reflector_v1_naive(x):
    # mesma função do item 1.c, mas com v1 = x1 - ||x|| calculado diretamente (sem reescrever)
    x = np.asarray(x)
    e1 = np.zeros(x.shape[0], dtype=x.dtype)
    e1[0] = 1

    x_norm = np.linalg.norm(x)
    y_norm2 = np.sum(x[1:] ** 2)

    v = x.copy()
    v[0] = x[0] - x_norm  # sujeita a cancelamento quando x1 ~ ||x||

    if x[0] >= 0:
        beta = (x_norm + x[0]) / (x_norm * y_norm2)
    else:
        beta = 1 / (x_norm * (x_norm - x[0]))

    return v, beta

# reaproveitando a trajetória x(s) -> ||x||e1 já construída no item 1.e
x0 = np.random.randint(1000.0, size=10)
xs, ss = calc_trajectoria(x0, 100, 14)

v1_naive = []
v1_estavel = []
for x in xs:
    v_n, _ = reflector_v1_naive(x)
    v_e, _ = reflector(x)  # usa a forma estável
    v1_naive.append(v_n[0])
    v1_estavel.append(v_e[0])

v1_naive = np.array(v1_naive)
v1_estavel = np.array(v1_estavel)

# erro relativo de cada versão de v1 frente ao valor "exato" (Float64 de referência)
v1_exato = v1_estavel  # ambas calculadas em float64; a estável não sofre cancelamento

erro_naive = np.abs(v1_naive - v1_exato) / np.abs(v1_exato)
print("Maior erro relativo (forma ingenua):", np.nanmax(erro_naive))
print("v1 (ingenuo) perto de s->0:", v1_naive[-1])
print("v1 (estavel) perto de s->0:", v1_estavel[-1])
````

````python
def calc_v1s(xs, dtype_a, dtype_b):
    results = []
    for x in xs:
        xA = x.astype(dtype_a)
        xB = x.astype(dtype_b)

        v_naive_A, _ = reflector_v1_naive(xA)
        v_naive_B, _ = reflector_v1_naive(xB)
        v_est_A, _ = reflector(xA)
        v_est_B, _ = reflector(xB)

        results.append({'v1_naive_A': float(v_naive_A[0]), 'v1_naive_B': float(v_naive_B[0]),
                        'v1_est_A': float(v_est_A[0]), 'v1_est_B': float(v_est_B[0])})
    return results

results_v1 = calc_v1s(xs, np.float32, np.float64)
r = pd.DataFrame(results_v1)

diff_naive = np.maximum(np.abs(r['v1_naive_A'] - r['v1_naive_B']) / np.abs(r['v1_naive_B']), 1e-16)
diff_est = np.maximum(np.abs(r['v1_est_A'] - r['v1_est_B']) / np.abs(r['v1_est_B']), 1e-16)

plt.figure(figsize=(8, 5))
sns.lineplot(x=ss, y=diff_naive, label=r"$v_1 = x_1 - \|x\|$ (ingenua)")
sns.lineplot(x=ss, y=diff_est, label=r"$v_1 = -\|y\|^2/(x_1+\|x\|)$ (estavel)")

plt.gca().invert_xaxis()
plt.xscale('log')
plt.yscale('log')
plt.xlabel(r"$s$ - proximidade $\|x\|e_1$ conforme $s\to 0 $")
plt.ylabel(r"erro relativo de $v_1$")
plt.title(r"Condicionamento relativo de $v_1$: forma ingenua vs estavel")

plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
````

O gráfico confirma a análise teórica: a forma ingênua $v_1 = x_1 - \|x\|$ perde precisão relativa de forma crescente conforme $s \to 0$ (isto é, conforme $x$ se aproxima de $\|x\|e_1$), chegando a ter erro relativo próximo de $1$ (perda total de algarismos significativos) nos casos mais extremos, já que `Float32` e `Float64` discordam cada vez mais quando o cancelamento se intensifica. Já a forma estável $v_1 = -\|y\|^2/(x_1+\|x\|)$ mantém o erro relativo próximo da precisão de máquina do `Float32` ($\sim 10^{-7}$) em todo o intervalo testado, sem degradar conforme $s \to 0$.

Isso confirma que $v_1$, calculado pela fórmula ingênua, é relativamente mal-condicionado quando $x$ está quase alinhado com $e_1$, mesmo que $v$ como vetor completo seja bem-condicionado ([item 1.d](#item-1d)). A fórmula alternativa, que evita a subtração de quantidades quase iguais, resolve esse problema reescrevendo $v_1$ apenas em termos de $\|y\|^2$ e $x_1 + \|x\|$, sem nunca calcular $\|x\| - x_1$ diretamente — exatamente a mesma estratégia usada para estabilizar $\beta$ no [item 1.b](#item-1b).
