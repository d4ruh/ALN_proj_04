**a) Autovetores.** Seja $A(\omega) = \begin{bmatrix} 0 & 1 \\ -\omega^2 & 0 \end{bmatrix}$. Faça um gráfico dos autovalores para $\omega$ variando de 0 a 1, e também dos autovetores. O que acontece quando $\omega \to 0^+$?


#parte computacional, gráfiquinho

... vemos isso porque os autovalores são $$\lambda = \pm i\omega$$ 


Cálculo:

equação característica $\det(A - \lambda I) = 0$.

$$A - \lambda I = \begin{bmatrix} 0 - \lambda & 1 \\ -\omega^2 & 0 - \lambda \end{bmatrix} = \begin{bmatrix} -\lambda & 1 \\ -\omega^2 & -\lambda \end{bmatrix}$$

Calculando o determinante:


$$\det(A - \lambda I) = (-\lambda)(-\lambda) - (1)(-\omega^2)$$

$$\det(A - \lambda I) = \lambda^2 + \omega^2$$


Igualando o determinante a zero para encontrar os autovalores:


$$\lambda^2 + \omega^2 = 0$$

$$\lambda^2 = -\omega^2$$

$$\lambda = \pm \sqrt{-\omega^2}$$

$$\lambda = \pm i\omega$$

Os autovalores são:

* $\lambda_1 = i\omega$
* $\lambda_2 = -i\omega$
Quando $\omega = 0$, ambos os autovalores são ($\lambda_1 = \lambda_2 = 0$).
